```python
# Dependencies and Setup
import pandas as pd

# File to Load (student data file and school data file)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="inner", on=["school_name", "school_name"])
school_data_complete.head(5)
```

```python
per_school_count = school_data_complete["school_name"].value_counts()
```

```python
budget = school_data_complete.groupby(["school_name"]).mean()["budget"]
```

```python
per_capital = budget / per_school_count
```

```python
average_math_scores = school_data_complete.groupby(["school_name"]).mean()["math_score"]
```

```python
average_reading_scores = school_data_complete.groupby(["school_name"]).mean()["reading_score"]
```

```python
math_score_df = school_data_complete[school_data_complete["math_score"]>=70]
reading_score_df = school_data_complete[school_data_complete["reading_score"]>=70]
overall_passing_df =school_data_complete[(school_data_complete["reading_score"]>=70) & (school_data_complete["math_score"]>=70) ]
```

```python
per_school_passing_math = math_score_df.groupby(["school_name"]).count()["student_name"] / per_school_count * 100
```

```python
per_school_passing_reading = reading_score_df.groupby(["school_name"]).count()["student_name"] / per_school_count * 100
```

```python
overall_passing_per_school = overall_passing_df.groupby(["school_name"]).count()["student_name"] / per_school_count * 100
```

```python
# Place all of the data found into a summary DataFrame
summary_df_by_school = pd.DataFrame({ "School type": school_type, 
                            "Total Student": per_school_count,
                           "Total school budget":budget,
                           "Per student budget":per_capital,
                           "Average math score":average_math_scores ,
                           "Average reading score":average_reading_scores,
                            "% passing math" : per_school_passing_math,
                            "% passing reading":per_school_passing_reading ,
                           "% overall passing":overall_passing_per_school
                                                    })
summary_df_by_school
```
